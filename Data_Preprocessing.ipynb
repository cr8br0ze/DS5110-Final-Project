```python
### WVP Data Collection - Phase I - M.xlsm
```

```python
all_df = []
xls = pd.ExcelFile("WVP Data Collection - Phase I - M.xlsm")
for sheet in xls.sheet_names:
    df = xls.parse(sheet)
    all_df.append(df)
index = 0
all_df[index].columns = all_df[index].iloc[0]
all_df[index] = all_df[index].drop(all_df[index].index[0])
all_df[index] = all_df[index].reset_index(drop=True)
```

```python
all_df[0] = all_df[0][:75]
```

```python
all_df[0] = all_df[0].iloc[:, [2,3,6,7,9,10, 11, 12]]
```

```python
real_df.append(all_df[0])
```

```python
real_df[6].head(3)
```

```python
df = pd.concat(real_df, ignore_index=True)
```

```python
df['Job Role'] = df[
    ['Occupational Category of Person Affected',
     'Occupational category of Person Impacted',
     'Occupational Category of Person Impacted:']
].bfill(axis=1).iloc[:, 0]

df['Assault Description'] = df[
    ['Primary Assault Description',
     'Primary Assault Description ',
     'Description of the Assault (Select All that Apply)']
].bfill(axis=1).iloc[:, 0]

df['Psych Impact'] = df[
    ['Emotional and/ or Psychological Impact',
     'Emotional and/or Psychological Impact']
].bfill(axis=1).iloc[:, 0]

df = df.drop(columns=[
    'Occupational Category of Person Affected',
    'Occupational category of Person Impacted',
    'Occupational Category of Person Impacted:',
    'Primary Assault Description',
    'Primary Assault Description ',
    'Description of the Assault (Select All that Apply)',
    'Emotional and/ or Psychological Impact',
    'Emotional and/or Psychological Impact'
])
```

```python
df.to_csv("cleaned.csv", index=False)
```
